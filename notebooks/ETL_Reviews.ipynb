```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import json as js
import ast as ast
import re
import json
```

```python

#Convertir el archivo 'australian_user_reviews.json' en dataframe

list_rev = []
usuarios_reviews = r'australian_user_reviews.json'
with open(usuarios_reviews, encoding='utf-8') as file:
    for line in file.readlines():
        list_rev.append(ast.literal_eval(line))

DF_Reviews = pd.DataFrame(list_rev)
DF_Reviews = DF_Reviews.explode('reviews',  ignore_index=True)
```

```python
Reviews2 = pd.DataFrame(list_rev)
Reviews = Reviews2.explode('reviews', ignore_index=True)
Reviews
```

```python
Reviews_expanded = pd.json_normalize(Reviews['reviews'])
Reviews_expanded
```

```python
Reviews_expanded=Reviews_expanded.replace('',None)
Reviews_expanded
```

```python
Reviews = Reviews.join(Reviews_expanded)
Reviews
```

```python
Reviews.drop(columns=['reviews','last_edited','user_url'],inplace=True)
Reviews
```
