```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import json as js
import ast as ast
import re
```

```python
#Convertir el archivo 'australian_users_items.json' en dataframe

list_items = []
australian_items = r'australian_users_items.json'
with open(australian_items, encoding='utf-8') as file2:
    for line2 in file2.readlines():
        list_items.append(ast.literal_eval(line2))

DF_Items = pd.DataFrame(list_items)
DF_Items = DF_Items.explode('items', ignore_index=True)
```

```python
# Normalizar los diccionarios en columnas separadas
Columnas_items = pd.json_normalize(DF_Items['items'])

# Concatenar el DataFrame original sin la columna de diccionario con el nuevo DataFrame
DF_Items = pd.concat([DF_Items.drop(columns=['items']), Columnas_items], axis=1)

# Mostrar el resultado
DF_Items
```

```python

#Eliminamos las filas en donde todos los valores sean nulos
DF_Items = DF_Items.dropna(how='all')
print(DF_Items)
```

```python
#Em caso de que hayan datos duplicados, los borramos
DF_Items= DF_Items.drop_duplicates(subset=['user_id', 'item_id'])
```

```python
#Borramos la ultima fila ya que son datos que no aportan
DF_Items = DF_Items.drop(index=5170014)
```
